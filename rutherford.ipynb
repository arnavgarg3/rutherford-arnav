```python
import matplotlib.pyplot as plt
import numpy as np
import math


def rford_fct(rad, E):
    k = 8.99 * 10**9  
    q = 1.6 * 10**-19 
    Q = 1.6 * 10**-19  
    dcs = ((k * q * Q) / (4 * E * (np.sin(rad/2))**2))**2

    return dcs


"""
dcs vs theta
"""
x1_values = np.linspace(math.pi/4, math.pi, 100)  
for i in range(6, 30, 5):
    y1_values = rford_fct(x1_values, i)
    plt.plot(x1_values, y1_values, label='E = ' + str(i) + " kJ")

plt.xlabel('Scattering Angle (rad)')
plt.ylabel('Differential Cross Section (barns/sr)')
plt.title('Differental Scattering Cross Section vs Scattering Angle')
plt.legend()
plt.show()

"""
dcs vs E
"""
x2_values = np.linspace(1, 10, 100)  
for i in range(1, 4, 1):
    y2_values = rford_fct(i, x2_values)
    plt.plot(x2_values, y2_values, label='theta = ' + str(i) + " rad")

plt.xlabel('Particle Energy (kJ)')
plt.ylabel('Differential Cross Section (barns/sr)')
plt.title('Differential Scattering Cross Section vs Incident Particle Energy')
plt.legend()
plt.show()












```
